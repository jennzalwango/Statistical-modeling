Decision tree and the Random Forest algorithms

```python

# Implement the descion tree and random forest on Iris data
# Four features (Sepal, length, sepal width, petal lenhth and petal width)
#Import libraries
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn. tree import plot_tree

#import for random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

import numpy as np

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
#Load the dataset(iris dataset), in sklearn
iris = load_iris()
x = iris.data #features
y = iris.target #target

feature_names = iris.feature_names
target_names = iris.target_names
```

```python
#split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)
```

```python
#Decision tree model
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
dt_pred = dt_model.predict(x_test)
```

```python
#Random forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
rf_pred = rf_model.predict(x_test)
```

```python
#evaluate the decision tree
print("Decision Tree Accuracy:", accuracy_score(y_test, dt_pred)) #accurancy
print("Decision Tree Classification Report:/n", classification_report(y_test, dt_pred, target_names=target_names)) #classification report
```

```python
#Confusion Matrix for Random Forest
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, rf_pred), annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Random Forest Confusion Matrix')
plt.show()
```

```python
#Feature Importance
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.barh(feature_names, dt_model.feature_importances_)
plt.title('Decision Tree Feature Importance')

plt.subplot(1, 2, 2)
plt.barh(feature_names, rf_model.feature_importances_)
plt.title('Random Forest Feature Importance')
plt.tight_layout()
plt.show()
```

```python
#visualize decision tree
plt.figure(figsize=(12, 8))
plot_tree(dt_model, feature_names=feature_names, class_names=target_names, filled=True)
plt.title('Decision Tree Structure')
plt.show()
#
```
