```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df.isnull().sum()
```

```python
plt.scatter(x=df['Year'], y=df['Per capita income (E)'], marker='+', color='blue')
plt.xlabel('Year')
plt.ylabel('Per capita income (E)')
plt.show()
```

```python
reg = LinearRegression()
reg.fit(df[['Year']], df['Per capita income (E)'])
```

```python
print("Predicted per captita income for 2027:", reg.predict([[2010]]))
```

```python
plt.scatter(x=df['Year'], y=df['Per capita income (E)'], marker='+', color='blue')
plt.plot(df['Year'], reg.predict(df[['Year']]), color='green')
plt.xlabel('Year')
plt.ylabel('Per capita income (E)')
plt.show()
```

```python
print(reg.coef_)
print(reg.intercept_)
```

# Multiple Linear Regression

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Activity 2 MLR.csv')
```

```python
df.head()
```

```python
# Define independent (X) and dependent (y) variables
X = df[['age', 'experience']]  # Independent variables
y = df['income']  # Dependent variable
```

```python
reg = LinearRegression()
reg.fit(X, y)
```

```python
new_values = [[43,12]]
predicted_income = reg.predict(new_values)
print("Predicted income:", predicted_income)
```
